```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
```

# Definig the Antecedent and Consequent

```python
Dengue=ctrl.Antecedent(np.arange(0,101,1),'Dengue Symptoms')
Fever=ctrl.Antecedent(np.arange(96,106,1),'Fever')
JMpain=ctrl.Antecedent(np.arange(0,11,1),'Joint Muscle Pain')
Ihead=ctrl.Antecedent(np.arange(0,11,1),'Headache especially with Eye Pain')
```

# Defining the descripters of Input and Output variables

```python
Dengue['Normal']=fuzzy.trapmf(Dengue.universe,[0,0,10,20])
Dengue['Incubation']=fuzzy.trimf(Dengue.universe,[10,30,40])
Dengue['Prodromal']=fuzzy.trimf(Dengue.universe,[30,55,70])
Dengue['Illness']=fuzzy.trapmf(Dengue.universe,[60,70,80,100])


Fever['Normal']=fuzzy.trapmf(Fever.universe,[96,96,98,99])
Fever['Moderate']=fuzzy.trimf(Fever.universe,[98,101,103])
Fever['High']=fuzzy.trapmf(Fever.universe,[102,103,104,105])

JMpain['Normal']=fuzzy.trapmf(JMpain.universe,[0,0,2,4])
JMpain['Moderate']=fuzzy.trimf(JMpain.universe,[3,5,7])
JMpain['High']=fuzzy.trapmf(JMpain.universe,[6,7,8,10])

Ihead['Normal']=fuzzy.trapmf(Ihead.universe,[0,0,2,4])
Ihead['Moderate']=fuzzy.trimf(Ihead.universe,[3,5,7])
Ihead['High']=fuzzy.trapmf(Ihead.universe,[6,7,8,10])
```

# Forming A Rule Base

```python
r1=ctrl.Rule(Fever['High'] & JMpain['High'] & Ihead['High'] , Dengue['Illness'])
r2=ctrl.Rule(Fever['High'] & JMpain['High'] & Ihead['Moderate'] , Dengue['Illness'])
r3=ctrl.Rule(Fever['High'] & JMpain['High'] & Ihead['Normal'] , Dengue['Prodromal'])
r4=ctrl.Rule(Fever['High'] & JMpain['Moderate'] & Ihead['High'] , Dengue['Illness'])
r5=ctrl.Rule(Fever['High'] & JMpain['Moderate'] & Ihead['Moderate'] , Dengue['Prodromal'])
r6=ctrl.Rule(Fever['High'] & JMpain['Moderate'] & Ihead['Normal'] , Dengue['Prodromal'])
r7=ctrl.Rule(Fever['High'] & JMpain['Normal'] & Ihead['High'] , Dengue['Prodromal'])
r8=ctrl.Rule(Fever['High'] & JMpain['Normal'] & Ihead['Moderate'] , Dengue['Prodromal'])
r9=ctrl.Rule(Fever['High'] & JMpain['Normal'] & Ihead['Normal'] , Dengue['Incubation'])
r10=ctrl.Rule(Fever['Moderate'] & JMpain['High'] & Ihead['High'] , Dengue['Prodromal'])
r11=ctrl.Rule(Fever['Moderate'] & JMpain['High'] & Ihead['Moderate'] , Dengue['Prodromal'])
r12=ctrl.Rule(Fever['Moderate'] & JMpain['High'] & Ihead['Normal'] , Dengue['Prodromal'])
r13=ctrl.Rule(Fever['Moderate'] & JMpain['Moderate'] & Ihead['High'] , Dengue['Prodromal'])
r14=ctrl.Rule(Fever['Moderate'] & JMpain['Moderate'] & Ihead['Moderate'] , Dengue['Incubation'])
r15=ctrl.Rule(Fever['Moderate'] & JMpain['Moderate'] & Ihead['Normal'] , Dengue['Incubation'])
r16=ctrl.Rule(Fever['Moderate'] & JMpain['Normal'] & Ihead['High'] , Dengue['Incubation'])
r17=ctrl.Rule(Fever['Moderate'] & JMpain['Normal'] & Ihead['Moderate'] , Dengue['Incubation'])
r18=ctrl.Rule(Fever['Moderate'] & JMpain['Normal'] & Ihead['Normal'] , Dengue['Incubation'])
r19=ctrl.Rule(Fever['Normal'] & JMpain['High'] & Ihead['High'] , Dengue['Prodromal'])
r20=ctrl.Rule(Fever['Normal'] & JMpain['High'] & Ihead['Moderate'] , Dengue['Prodromal'])
r21=ctrl.Rule(Fever['Normal'] & JMpain['High'] & Ihead['Normal'] , Dengue['Incubation'])
r22=ctrl.Rule(Fever['Normal'] & JMpain['Moderate'] & Ihead['High'] , Dengue['Prodromal'])
r23=ctrl.Rule(Fever['Normal'] & JMpain['Moderate'] & Ihead['Moderate'] , Dengue['Incubation'])
r24=ctrl.Rule(Fever['Normal'] & JMpain['Moderate'] & Ihead['Normal'] , Dengue['Normal'])
r25=ctrl.Rule(Fever['Normal'] & JMpain['Normal'] & Ihead['High'] , Dengue['Incubation'])
r26=ctrl.Rule(Fever['Normal'] & JMpain['Normal'] & Ihead['Moderate'] , Dengue['Normal'])
r27=ctrl.Rule(Fever['Normal'] & JMpain['Normal'] & Ihead['Normal'] , Dengue['Normal'])
```

```python
rules = [r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12,r13,r14,r15,r16,r17,r18,r19,r20,r21,r22,r23,r24,r25,r26,r27]
Diagnosing = ctrl.ControlSystem(rules)
diag = ctrl.ControlSystemSimulation(Diagnosing)
```

# Passing input to the Control System and Defuzzifized Output

```python
diag.input['Fever'] = 105
diag.input['JMpain'] = 9
diag.input['Ihead'] = 9
diag.compute()
print(diag.output['Dengue'])
```
